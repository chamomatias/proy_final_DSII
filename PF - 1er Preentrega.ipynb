##4.4 *Outliers* en las variables num茅ricas

No siempre se deben eliminar los *outliers* porque dependiendo de la variable num茅rica analizada estos pueden contener informaci贸n importante.

Creemos gr谩ficas tipo "boxplot" de las columnas num茅ricas:

```python
# Ver columnas num茅ricas del dataset original
cols_num = df.select_dtypes(include='number').columns
print(cols_num)
```

```python
# Lista de columnas num茅ricas
cols_num = [

            'Age',
            'BMI',
            'HBA1C',
            'Fasting_Blood_Sugar',
            'Postprandial_Blood_Sugar'
            ]

# Funci贸n para contar outliers usando IQR
def contar_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    return len(outliers)

# 1. Mostrar cantidad de outliers por columna
print(" Outliers detectados por variable:\n")
for col in cols_num:
    n_outliers = contar_outliers(data, col)
    print(f"{col}: {n_outliers} outliers")

# 2. Gr谩ficos boxplot de cada variable
fig, ax = plt.subplots(nrows=len(cols_num), ncols=1, figsize=(10, 4*len(cols_num)))
fig.subplots_adjust(hspace=0.6)

for i, col in enumerate(cols_num):
    sns.boxplot(x=data[col], ax=ax[i], color='skyblue', fliersize=5)
    ax[i].set_title(f'{col} (outliers en puntos)')
    ax[i].set_xlabel('Valor')

plt.tight_layout()
plt.show()
```

**Observaciones:**
- No se detectaron outliers.
